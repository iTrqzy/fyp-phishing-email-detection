## 1) Project in one sentence

I built an **interpretable phishing email detection system** that combines **email text features** with **metadata-based structural indicators** to see whether a hybrid approach can improve detection quality and make decisions more defensible.

**Quick demo line:**  
> “My project was about testing whether combining what an email says with how it is structured gives a better and more reliable phishing detector than using text alone.”

## 2) Why I chose this topic

- Phishing is still a major cybersecurity problem.
- Modern phishing emails are becoming more convincing, especially with **LLM-generated language**.
- That makes it risky to rely only on simple keyword checks or obvious spelling mistakes.
- I wanted a solution that was not just accurate, but also **explainable** and more useful in a realistic security setting.

**Quick demo line:**  
> “I chose this because phishing is evolving. Attackers can now produce very realistic emails, so I wanted to test a more robust and interpretable approach than basic text-only filtering.”

## 3) Main research that shaped the project

### Olea et al. (2025)
- Showed that people struggle more with **LLM-generated phishing**.
- This supported my view that phishing text is getting harder to detect using obvious surface clues.

### Krause et al. (2019)
- Showed that **metadata/header information** can be very powerful on its own.
- This justified building a metadata-based stage rather than only focusing on the email body.

### Mittal et al. (2022)
- Supported the idea that **hybrid / multi-feature systems** can outperform single-feature approaches.
- This helped justify my final hybrid design.

### Gholampour and Verma (2023)
- Showed that NLP phishing models can be weakened by small adversarial edits.
- This is why I included robustness-oriented evaluation rather than relying only on one clean test set.

### Seidenberger et al. (2025)
- Reinforced the value of **trust signals** such as sender authentication and source categorisation.
- This influenced my thinking around metadata and operational decision-making.

**Quick demo line:**  
> “The literature pushed me toward a staged hybrid design: strong text modelling, metadata as a second evidence source, and then fusion with interpretability.”

## 4) What my report was trying to answer

### Main aim
To design and evaluate an **interpretable hybrid phishing detector** that combines TF-IDF text modelling with metadata-derived indicators, and test whether this improves detection performance and probability reliability compared with a strong text-only baseline.

### In simpler words
My report was basically asking:

> “Does combining text and metadata give me a better phishing detector than text alone, and does it also make the output more trustworthy?”

**Quick demo line:**  
> “The project was not just about getting a higher score. It was also about whether the model’s decisions become more reliable and easier to justify.”

## 5) Why I used this methodology

I used a **quantitative experimental methodology** because I needed to compare different model designs under controlled conditions.

I measured:
- accuracy
- precision
- recall
- F1-score
- ROC-AUC
- PR-AUC
- confusion-matrix errors
- calibration / probability reliability
- runtime
- threshold behaviour
- robustness under more difficult phishing inputs

### Why this was suitable
Because the project question is comparative:
- text only vs metadata only vs hybrid
- same dataset split
- same test rows
- measurable outcomes

**Quick demo line:**  
> “I used an experimental methodology because I wanted controlled, evidence-based comparison rather than opinion-based claims.”

## 6) What I actually built

### Stage 1 — Text-only model
- TF-IDF vectorisation
- Logistic Regression
- Purpose: establish a strong baseline using the email content

### Stage 2 — Metadata-only model
- Engineered structural / header-derived indicators
- Purpose: test whether metadata alone provides useful phishing evidence

### Stage 3 — Hybrid model
- Stacking approach
- Combines Stage 1 and Stage 2 probability outputs
- Purpose: fuse both evidence streams in an interpretable way

### Extra design decisions
- consistent train/test split across all stages
- reproducibility controls
- CLI prediction workflow
- explanation mode to support transparency

**Quick demo line:**  
> “I deliberately built the project in stages so I could prove what each evidence source contributes before combining them.”

## 7) Main findings from the report

### Core result
- **Stage 1** was already a very strong baseline.
- **Stage 2** showed that metadata alone had value, but it was not strong enough as a standalone detector.
- **Stage 3** gave the **best overall balance** and the **most reliable probabilities**.

### Reported hybrid results
- Accuracy: **0.977**
- Precision: **0.972**
- Recall: **0.966**
- F1-score: **0.969**
- ROC-AUC: **0.997**
- PR-AUC: **0.995**

### What this means
The hybrid model did improve on the text-only approach, but the improvement was **modest**, not dramatic.  
That is actually an honest and important finding:
- text already captured a lot of the useful signal
- metadata still added value
- the biggest gain was not only raw classification, but **better probability reliability and decision defensibility**

**Quick demo line:**  
> “My key finding is that hybrid fusion helped, but in a qualified way. The score increase was small, yet the outputs became more defensible and better suited to threshold-based triage.”

## 8) What I learned from the results

### 1. Text is still very powerful
A simple classical NLP pipeline can still perform extremely well on phishing email classification when the dataset is controlled.

### 2. Metadata is useful, but limited on its own
Metadata helped, especially as supporting evidence, but it struggled as a standalone model because the available dataset did not always contain rich real-world trust signals.

### 3. Calibration matters
A model is not only about correct labels. In a security setting, the **confidence score** also matters because it affects triage, quarantine, and analyst trust.

### 4. Hybrid value is practical, not just numerical
Even where headline gains are small, combining evidence streams can make the system more operationally useful.

**Quick demo line:**  
> “The project taught me that better security models are not only about higher accuracy, but also about how reliable and explainable their confidence is.”

## 9) Honest limitations from the report

- Evaluation was based on a **single dataset split**
- The metadata stream used **engineered proxy indicators**, not a full enterprise email-authentication pipeline
- The dataset did not always contain richer trust signals such as full SPF/DKIM/DMARC context
- Robustness testing was useful, but still limited compared with real attacker behaviour
- So I should be careful not to overclaim real-world generalisation

**Quick demo line:**  
> “One of my main reflections is that strong results on one dataset do not automatically prove deployment readiness, especially in security.”

## 10) Future work

The report points toward:
- richer header-authentication features
- broader cross-dataset validation
- more realistic robustness testing
- stronger deployment realism
- improved trust-signal extraction
- extending the triage-oriented use of calibrated probabilities

**Quick demo line:**  
> “If I continued the project, I would focus on richer real-world trust signals and broader validation, because that is where the current prototype is still limited.”

## 11) Good short answers for likely demo questions

### Why did you do this project?
Because phishing is still a serious problem, and modern phishing emails are becoming more realistic. I wanted to test whether combining text and metadata could produce a detector that is both effective and easier to justify.

### Why not only use text?
Text is powerful, but it can be manipulated. Metadata provides a different evidence source, so combining both can reduce reliance on one channel.

### Why not only use metadata?
Metadata alone missed too much phishing in my experiments. It worked better as supporting evidence than as a complete solution.

### What is your main conclusion?
The hybrid approach slightly improved performance, but more importantly it improved probability reliability and made the decision process more defensible.

### Was the project successful?
Yes. I met the main aim of designing, implementing, and evaluating a staged hybrid phishing detector, and I produced evidence showing where hybrid fusion helps and where its current limits are.

### What are you most proud of?
The staged design and evaluation, because it lets me justify each modelling step rather than jumping straight to a black-box final model.
